线性回归损失函数$$Loss = \frac {1}{n} \sum_{i=1}^n \frac{1}{2} (\hat y^{(i)} - y^{(i)})^2$$$$\hat y^{(i)} = x_1^{(i)}w_1 + x_2^{(i)}w_2$$
加入L2范数惩罚项后$$L2Loss = Loss + \frac {\lambda}{2n} ||w||^2$$

$lambda$越大，惩罚项在损失函数内占比越大，$||w||^2 = w_1^2 + w_2^2$越接近0。在nimibatch gd中，原始的参数更新：
$$w_1 \leftarrow w_1 - \frac {\eta}{|B|}\sum_{i \in B}x_1^{(i)}(x_1^{(i)}w_1 + x_2^{(i)}w_2 + b - y^{(i)})$$
新的参数更新公式为：
$$w_1 \leftarrow (1-\frac {\eta \lambda}{|B|})w_1 - \frac {\eta}{|B|}\sum_{i \in B}x_1^{(i)}(x_1^{(i)}w_1 + x_2^{(i)}w_2 + b - y^{(i)})$$

# 验证L2正则化（权重衰减)应对过拟合

```python
%matplotlib inline
import torch
import torch.nn as nn
import numpy as np
import sys
sys.path.append("..") 
import d2lzh_pytorch as d2l

n_train, n_test, num_inputs = 20, 100, 200

real_w = torch.ones(num_inputs) * 0.01
real_b = 0.05

features = torch.randn((n_train + n_test), num_inputs) # (120,200)
labels = torch.matmul(features, real_w) + real_b
labels += torch.tensor(np.random.normal(0,0.01, size=labels.size()), dtype = torch.float) # 添加噪音
train_features, test_features = features[:n_train, :], features[n_train:, :]
train_labels, test_labels = labels[:n_train], labels[n_train:]
```

```python
# 初始化参数，为每个参数都附上梯度
def init_params():
    w = torch.randn((num_inputs, 1), requires_grad=True)
    b = torch.zeros(1, requires_grad=True)
    return [w, b]

# 定义L2范数惩罚项
def L2_penalty(w):
    return (w**2).sum() / 2

def L1_penalty(w):
    return w.sum() / 2
```

```python
batch_size = 1
num_epochs = 100
lr = 0.003

model, loss_func = d2l.linreg, d2l.squared_loss

dataset = torch.utils.data.TensorDataset(train_features, train_labels)
train_iter = torch.utils.data.DataLoader(dataset, batch_size, shuffle=True)

def fit_and_plot(lambd, penalty = "L2"):
    w, b = init_params()
    train_loss, test_loss = [],[]
    for _ in range(num_epochs):
        for x, y in train_iter:
            y_hat = model(x,w,b)
            if penalty == "L2":
                loss = loss_func(y_hat, y) + lambd * L2_penalty(w)
            else:
                loss = loss_func(y_hat, y) + lambd * L1_penalty(w)
            loss = loss.sum()
            
            if w.grad is not None:
                w.grad.data.zero_()
                b.grad.data.zero_()
            loss.backward()
            d2l.sgd([w,b], lr, batch_size)
        train_loss.append(loss_func(model(train_features, w, b), train_labels).mean().item())
        test_loss.append(loss_func(model(test_features, w, b), test_labels).mean().item())
        
    d2l.semilogy(range(1, num_epochs + 1), train_loss, 'epochs', 'loss',
                 range(1, num_epochs + 1), test_loss, ['train', 'test'])
    print('L2 norm of w:', w.norm().item())
```

```python
fit_and_plot(lambd=0)
features       
```

```python
fit_and_plot(lambd=3)
features
```

```python
fit_and_plot(lambd=5)
features
```
