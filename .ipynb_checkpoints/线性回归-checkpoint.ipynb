```python
# %matplotlib inline # 设置为嵌入显示
import torch
from IPython import display
from matplotlib import pyplot as plt
import numpy as np
import random
from d2lzh_pytorch.utils import use_svg_display, set_figsize
```

# 构造一个简单地人工数据集

```python
input_dim = 2
input_num = 1000
real_w = [2, -3.4]
real_b = 4.2
features = torch.randn(input_num, input_dim, dtype=torch.float32) # (1000,2)
features[:3]
```

```python
labels = real_w[0] * features[:,0] + real_w[1] * features[:,1] + real_b
# 加噪声
labels += torch.tensor(np.random.normal(0, 0.01, size = labels.size())
                      , dtype = torch.float32)
labels[:3]
```

```python
set_figsize()
plt.scatter(features[:,1].numpy(), labels.numpy(), 1)
```

# 读取数据

```python
def data_iter(batch_size, features, labels):
    num_samples = features.shape[0]
    indices = list(range(num_samples))
    random.shuffle(indices)
    for i in range(0, num_samples, batch_size):
        j = torch.LongTensor(indices[i:min(i + batch_size, num_samples)])
        yield features.index_select(0, j), labels.index_select(0, j)
```

```python
batch_size = 10
next(data_iter(batch_size, features, labels))
```

# 从0实现线性回归
# 初始化模型参数

```python
w = torch.tensor(np.random.normal(0,0.01,(input_dim, 1)),
                 dtype = torch.float32)
b = torch.zeros(1, dtype = torch.float32)

# 需要求梯度，所以设置
w.requires_grad_(requires_grad=True)
b.requires_grad_(requires_grad=True)
```

# 定义模型, loss， optimizer

```python
def linreg(X,w,b):
    return torch.mm(X,w) + b

def squared_loss(y_hat, y):
    # 返回的是向量，且除以了2
    return (y_hat - y.view(y_hat.size())) ** 2 / 2
            
def sgd(params, lr, batch_size):
    # params 是所有trainable的参数
    for param in params:
        # 注意这里改的是data
        param.data -= lr * param.grad / batch_size
            
```

```python
# 设置超参数
lr = 0.03
EPOCH = 10
model = linreg
loss = squared_loss

for epoch in range(EPOCH):
    for X, y in data_iter(batch_size, features, labels):
        l = loss(model(X, w, b), y).mean()
        l.backward()
        # 优化，更新param
        sgd([w, b], lr, batch_size)
        
        # 梯度清零
        w.grad.data.zero_()
        b.grad.data.zero_()
        
    train_l = loss(model(features, w, b), labels)
    print(f"epoch {epoch+1}, loss {train_l.mean()}")
```

```python
print(f"real w {real_w}  w {w.data}")
print(f"real b {real_b} b {b.data}")
```

# 利用pytorch简洁地实现线性回归

```python
# 读取数据
import torch.utils.data as Data
batch_size = 10
# 融合特征和标签
dataset = Data.TensorDataset(features, labels)
# 构建迭代器
data_iter = Data.DataLoader(dataset, batch_size, shuffle=True)
```

# 定义模型

```python
# 方法一
class LinearNet(torch.nn.Module):
    def __init__(self, n_feature):
        super(LinearNet, self).__init__()
        self.linear = torch.nn.Linear(n_feature,1)
    
    def forward(self,x):
        y = self.linear(x)
        return y
    
model = LinearNet(input_dim)
model
```

```python
# 方法二
model_1 = torch.nn.Sequential(
    torch.nn.Linear(input_dim, 1)
    )

# 方法三
model_2 = torch.nn.Sequential()
model_2.add_module('linear', torch.nn.Linear(input_dim,1))

# 方法四
from collections import OrderedDict
model_3 = torch.nn.Sequential(OrderedDict([
    ('linear', torch.nn.Linear(input_dim,1))
]))

print(model_3[0])
```
